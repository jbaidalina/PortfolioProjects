```python
import pandas as pd
import requests
import json

headers = {
    "accept": "application/json",
    "Authorization": "Bearer *****"
}

#pd.options.display.max_columns = 30
```

```python
#Importing the Movies Dataset
```

```python
# get the list of Franch movies releases in 1990-2009

base_url = "https://api.themoviedb.org/3/discover/movie?include_video=false&language=en&page={page}&primary_release_date.gte=1990-01-01&primary_release_date.lte=2009-12-31&with_original_language=fr&sort_by=primary_release_date.asc"
result = []
page_number = 1

while True:
    url = base_url.format(page = page_number)
    #print(url)
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        #print(page_number, response.json()["total_pages"])
        result.extend(response.json()["results"])
        if (page_number < response.json()["total_pages"]) & (page_number < 500):
            page_number += 1
        else:
            break
    else:
        print(f"Failed to retrieve data. Page {page_number} Status code: {response.status_code}")
        break
print(page_number)
print(response.json()["total_pages"])
movie_overview = pd.DataFrame(result)

```
